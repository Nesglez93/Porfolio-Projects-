```python
# I created a Columnm Called survival in Boolean 1-0
df['Survival'] = df.Survived.map({0 : 'Died', 1 : 'Survived'})
df.Survival.head()
survival_pass =df.groupby('Sex')["Survival"].value_counts(normalize = True).unstack()
```

```python
# 7 Mostrar por pantalla el porcentaje de personas que sobrevivieron en cada clase.
```

```python
print("Porcentaje de sobrevivientes de acuerdo a Clase")
porcentaje = (df.groupby(['Pclass']))['Survival'].value_counts(normalize = True).unstack()
```

```python
porcentaje
```

```python
# Set figure size
plt.rc('figure', figsize=(10, 5))
# Setup a figue of plots
porcentaje.plot(kind='bar',
               stacked=True)

plt.legend(('Fallecieron', 'Sobrevivieron'), loc='best')
plt.title('Sobrevivientes por Clase')
plt.xlabel('Clase')
plt.ylabel('Percentaje')
plt.xticks(rotation=1)

plt.show()
```
